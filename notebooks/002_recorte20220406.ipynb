```python
import pandas as pd
```

```python
f = "../data/raws/20220406/ESOLMET-CR1000_Tabla_diez_min.dat"
tmx = pd.read_csv( f, skiprows=[0,2,3], sep="\t", encoding="latin1", engine="python")
tmx.head()
```

```python
tmx["TIMESTAMP"] = pd.to_datetime(tmx["TIMESTAMP"], dayfirst=True,format="%d/%m/%Y %H:%M", errors="coerce")
tmx = tmx.set_index("TIMESTAMP")
tmx = tmx.sort_index()
tmx
```

```python
inicio = "2022-02-04 15:10"
fin    = "2022-04-06 10:50"

tmx_recorte = tmx.loc[inicio:fin]
tmx_recorte.head()
```

```python
output_path = "../data/raws/20220406_recorte.parquet"
tmx_recorte.to_parquet(output_path, index=True)
```
