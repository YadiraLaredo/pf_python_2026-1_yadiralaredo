```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import glob
import os
```

```python
df_1= pd.read_parquet("../data/raws/20220204_recorte.parquet")
df_2= pd.read_parquet("../data/raws/20220406_recorte.parquet")
df_3= pd.read_parquet("../data/raws/20220517_recorte.parquet")
df_4= pd.read_parquet("../data/raws/20221111_recorte.parquet")
```

```python
df_grandote = pd.concat([df_1, df_2, df_3, df_4], axis=0).sort_index()
```

```python
df_grandote.index = pd.to_datetime(df_grandote.index)
df_grandote = df_grandote.sort_index()
```

```python
df_grandote
```

```python
df_grandote.to_parquet("../data/processed/df_grandote.parquet", index=True)
```

```python
df_grandote.dtypes
```

```python
for col in ["I_glo_Avg", "I_dir_Avg", "I_dif_Avg", "WindDir"]:
    df_grandote[col] = (
        df_grandote[col]
        .astype(str)          
        .str.strip()         
        .str.replace(",", ".", regex=False)  
        .str.replace("—", "", regex=False)   
        .str.replace("-", "", regex=False))
```

```python
for col in ["I_glo_Avg", "I_dir_Avg", "I_dif_Avg", "WindDir"]:
    df_grandote[col] = pd.to_numeric(df_grandote[col], errors="coerce")
```

```python
df_grandote.to_parquet("../data/processed/df_concatenado.parquet", index=True)
```
