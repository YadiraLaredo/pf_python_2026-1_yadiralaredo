## Ese error es completamente normal cuando se intenta leer archivos .dat de estaciones tipo Campbell (CR1000): 
no están en UTF-8, sino casi siempre en Latin-1 (ISO-8859-1) y además suelen traer caracteres raros (°C, comas, tabs, encabezados dobles, etc.).

```python
f = "../data/raws/20220204/ESOLMET-CR1000_Tabla_diez_min.dat"
tmx = pd.read_csv( f, encoding="latin1", engine="python", sep="\t", on_bad_lines="skip")

tmx.info()
```

```python
f = "../data/raws/20220204/ESOLMET-CR1000_Tabla_diez_min.dat"
tmx = pd.read_csv( f, skiprows=[0,2,3], sep="\t", encoding="latin1", engine="python")
tmx.head()
```

```python
tmx.columns
```

```python
tmx["TIMESTAMP"] = pd.to_datetime(tmx["TIMESTAMP"], dayfirst=True, format="%d/%m/%Y %H:%M", errors="coerce")
tmx = tmx.set_index('TIMESTAMP')
tmx = tmx.sort_index()
tmx
```

### Dado que hay demasiados archivos con tantos datos tan diferentes:

Busqué manualmente en cada uno fechas que concatenar, hasta el punto de formar un df con un año lo mas completo posible, para el resto de libretas, busqué se recortaron las fechas adecuadas segun mi analisis.

```python
inicio = "2021-11-11 14:20"
fin    = "2022-02-04 15:00"

tmx_recorte = tmx.loc[inicio:fin]
tmx_recorte.tail()
```

```python
output_path = "../data/raws/20220204_recorte.parquet"
tmx_recorte.to_parquet(output_path, index=True)
```
