```python
import pandas as pd
```

```python
f = "../data/raws/20221111/ESOLMET-IP_Tabla_diez_min.dat"
tmx = pd.read_csv( f, skiprows=[0,2,3], sep=",", encoding="latin1", engine="c", on_bad_lines='skip' )
tmx
```

```python
tmx["TIMESTAMP"] = pd.to_datetime( tmx["TIMESTAMP"], format="%Y-%m-%d %H:%M:%S", errors="coerce")
tmx = tmx.set_index("TIMESTAMP")
tmx = tmx.sort_index()
tmx
```

```python
inicio = "2022-05-17 11:50"
fin    = "2022-11-11 14:20"

tmx_recorte = tmx.loc[inicio:fin]
tmx_recorte.head()
```

```python
output_path = "../data/raws/20221111_recorte.parquet"
tmx_recorte.to_parquet(output_path, index=True)
```
