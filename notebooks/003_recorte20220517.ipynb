```python
f = "../data/raws/20220517/ESOLMET-IP_Tabla_diez_min.dat"
tmx = pd.read_csv( f, skiprows=[0,2,3], sep=",", encoding="latin1", engine="c", on_bad_lines='skip' )
tmx
```

```python
# Ver ejemplos de TIMESTAMP
print(tmx["TIMESTAMP"].head(10))
```

```python
tmx["TIMESTAMP"] = pd.to_datetime( tmx["TIMESTAMP"], format="%Y-%m-%d %H:%M:%S", errors="coerce")
tmx = tmx.set_index("TIMESTAMP")
tmx = tmx.sort_index()
tmx
```

```python
inicio = "2022-04-22 15:00"
fin    = "2022-05-17 11:40"

tmx_recorte = tmx.loc[inicio:fin]
tmx_recorte.head()
```

```python
output_path = "../data/raws/20220517_recorte.parquet"
tmx_recorte.to_parquet(output_path, index=True)
```
